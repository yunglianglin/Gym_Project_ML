```python
import pandas as pd
#把語料讀出來
df_idf = pd.read_csv('./ppt_clean_30_wds_time.csv', encoding='utf-8')
```

```python
df_idf.head(10)
```

```python
len(df_idf.loc[3, 'article'])
```

```python
cmt01_18 = df_idf[df_idf.time.between('2018/1/1', '2018/1/31')].reset_index(drop=True)
cmt01_18 = cmt01_18.loc[0:, 'article']
```

```python
len(cmt01_18[2].split(" "))
```

```python
wordcount={}

for j in range (0, len(cmt01_18)):
    print(j)
    for x in cmt01_18[j].split(" "):
        if x not in wordcount:
            wordcount[x] = 1
        else:
            wordcount[x] += 1


import operator
sorted_x = sorted(wordcount.items(), key = operator.itemgetter(1), reverse = True)
sorted_x

import pandas as pd 
df = pd.DataFrame.from_dict(sorted_x)
```

```python
#設置pub_date的欄位, 並加入日期
df['pub_date'] = '2018-01-15'
#重新命名column
df.rename(columns={0: 'word', 1: 'freq'}, inplace=True)
#濾掉freq=1
df = df[df.freq.between(20, 100)]
df
```

```python
cmt_0118 = df_idf[df_idf.time.between('2018/1/1', '2018/1/31')].reset_index(drop=True)
cmt_0118 = cmt_0118.loc[0:, 'article']

wordcount={}
for j in range (0, len(cmt_0118)):
    print(j)
    for x in cmt_0118[j].split(" "):
        if x not in wordcount:
            wordcount[x] = 1
        else:
            wordcount[x] += 1


import operator
sorted_x = sorted(wordcount.items(), key = operator.itemgetter(1), reverse = True)
sorted_x

import pandas as pd 
df_0118 = pd.DataFrame.from_dict(sorted_x)

#設置pub_date的欄位, 並加入日期
df_0118['pub_date'] = '2018-01-15'
#重新命名column
df_0118.rename(columns={0: 'word', 1: 'freq'}, inplace=True)
#濾掉freq=1
df_0118 = df_0118[df_0118.freq.between(20, 100)]
df_0118
```
