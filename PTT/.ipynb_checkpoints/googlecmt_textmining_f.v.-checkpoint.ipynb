```python
#只取comment和rating
import pymongo
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017')
db = client.google_db
collection = db.google_col

results = []
#$ne selects the documents where the value of the field is not equal to the specified value
#選取comment 為非空值
for obj in collection.find({'comment':{'$ne':''}}, {"_id":0, "comment":1, "rating":1}):
      results.append(obj.copy())
#     results['comment'] = obj['comment'].copy()
#     results['rating'] = obj['rating'].copy()
#     results.append(obj.copy())  
results
```

# 計算詞頻

```python
from  sklearn.feature_extraction.text  import  CountVectorizer 
import pandas as pd
```

```python
df_cpus = pd.read_csv('./corpus_after_stpw.csv', encoding='utf-8')
#把語料讀出來
df_cpus_cmts = df_cpus.loc[0:, 'comment']
df_cpus.shape
```

```python
# 詞頻小於50的會被濾掉

vectorizer = CountVectorizer(token_pattern='(?u)\\b\\w+\\b',min_df =50)  
vectorizer
```

```python
output = vectorizer.fit_transform(df_cpus_cmts)
#output.toarray() #查看詞頻結果  
word = vectorizer.get_feature_names() #d看看我們的詞庫裡有什麼 
print(word)  
len(word) #581
```

```python
from  sklearn.feature_extraction.text  import  TfidfTransformer  
import pandas as pd
transformer = TfidfTransformer()  
 
#將詞頻矩陣X統計成TF-IDF值  
tfidf = transformer.fit_transform(output)  
#查看數據結構 tfidf[i][j]表示i類文本中的tf-idf權重  
tfs = tfidf.toarray()
tfs_df = pd.DataFrame(tfs, columns = word)
len(tfs)
```

```python
import pandas as pd
#把所有tf-idf的特徵變成所有dataframe的columns
fin = pd.concat([tfs_df, df], axis = 1)
#把所有features的值相加, 並新增一個欄位sum_feature_val
fin ['sum_feature_val'] = fin[word].sum(axis=1)
```

```python
fin
```

```python
fin.loc[32, 'sum_feature_val'] == 0
```

```python
#先抓出tf array 為 0 的index
zero_val = []
for k in range (0, len(fin)):
    if fin.loc[k, 'sum_feature_val'] == 0:
        zero_val.append(k)
print(len(zero_val))

#再把這些tf array的row清掉
fin_ad = fin.drop(fin.index[zero_val])

#把其中的幾列清掉後，其index還是不變，所以之後再根據index用for迴圈實會出錯
#例如index=32 這列已被清除, 當for跑到32時就會報錯, 因此必須reset index
#drop=true 表示把舊的index清除,換成新的
cpus_for_log = fin_ad.reset_index(drop=True)
print(len(cpus_for_log))

#把rating 的 5, 4, 3改成1, 2,1改成0 
cpus_for_log['rating'].replace([5,4,3,2,1], [1,1,0,-1,-1], inplace=True)

# import pandas as pd 
# n.to_csv('./cpus_for_log.csv', index=False, encoding='utf-8')
```

```python
cpus_for_log
```

```python
import numpy as np
#cpus_for_log['comment']的datatype是<class 'pandas.core.series.Series'>, 所以必須先轉成 numpy.ndarray
X = cpus_for_log.loc[0:, word].values
y = cpus_for_log.loc[0:, 'rating'].values
X.shape
y.shape
```

```python
#把資料分成 70%的training set, 30%的 testingset
# random_state=0 確保每次切分資料的結果都相同
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
#建立好一個LR model
text_model = LogisticRegression()
text_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

y_pred = text_model.predict(X_train)

print('MSE train: %.3f' % (
        mean_squared_error(y_train, y_pred)))
print('R^2 train: %.3f' % (
        r2_score(y_train, y_pred)))
```

```python
y_pred = text_model.predict(X_test)

print('MSE test: %.3f' % (
        mean_squared_error(y_test, y_pred)))
print('R^2 tset: %.3f' % (
        r2_score(y_test, y_pred)))
```
